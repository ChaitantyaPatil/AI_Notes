```python
text_file_path = r"data\text_files\machine_learning.txt" 
```

### TextLoader - Read Single File

```python
from langchain_community.document_loaders import TextLoader

#loading a single text file
loader = TextLoader(text_file_path,encoding="utf-8")

document = loader.load()

print(f"loaded {len(document)} documents")
print(f"content {document[0].page_content[:100]} ...")
print(f"metadata{document[0].metadata}")
```

### DirectoryLoader - multiple text files  

```python
from langchain_community.document_loaders import DirectoryLoader

#load all files from the directory

dir_loader = DirectoryLoader(
    "data/text_files",
    glob="**/*.txt", #pattern to match files
    loader_cls=TextLoader,
    loader_kwargs={'encoding':'utf-8'},
    show_progress=True
)

document= dir_loader.load()

print(f"loaded {len(document)} documents")
print(f"content {document[0].page_content[:100]} ...")
print(f"metadata{document[0].metadata}")
```

### Text Splitting Strategies

```python
### Method 1 - Character Text Splitter
text = document[0].page_content

char_splitter = CharacterTextSplitter(
    separator= "\n",
    chunk_size = 200, # max size of char
    chunk_overlap = 20, # overlap of char in chunks
    length_function =len # how to measure
)
char_chunks = char_splitter.split_text(text)

print(f"created {len(char_chunks)}")
print(f"first chunk: {char_chunks[0][:100]}")
print("*"*60)
for i in range(len(char_chunks)):
    print(char_chunks[i])
    print("_"*60)
```

```python



text = """
Artificial Intelligence (AI) is transforming the world.
It enables machines to learn from data, make decisions,
and solve complex problems. AI is used in healthcare,
finance, transportation, and many other fields.
"""

splitter = RecursiveCharacterTextSplitter(
    chunk_size=50,      # max characters per chunk
    chunk_overlap=10    # overlap between chunks
)

chunks = splitter.split_text(text)

for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}:")
    print(chunk)
    print("-" * 30)
```
